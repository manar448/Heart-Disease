```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load cleaned data
cleaned_df = pd.read_csv("D:\AI & ML Sprints\data\cleaned_heart_disease.csv")
X = cleaned_df.drop('target', axis=1)
y = cleaned_df['target']

# Apply PCA
pca = PCA(n_components=0.95)  # Retain 95% variance
pca_result = pca.fit_transform(X)

# Plot explained variance
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Variance Retention')
plt.grid(True)
plt.savefig('D:/AI & ML Sprints/results/pca_variance_plot.png')

# Save PCA-transformed data
pd.DataFrame(pca_result).to_csv('D:\AI & ML Sprints\data\pca_transformed_data.csv', index=False)
```
